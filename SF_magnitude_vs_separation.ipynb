```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
full_sample = pd.read_csv('/Users/Jess/sf_m_dwarfs/sf_data_files/final_sample_gm_binaries_flux2.csv')
```

```python
app_mag_pri = full_sample['Kepler']
app_mag_sec = full_sample['sec_mags']
mag_diff = app_mag_sec - app_mag_pri
orbit = full_sample['periods']
mass_pri = full_sample['Mact']
mass_sec = full_sample['sec_mass']
```

```python
numer = ((orbit * 86400)**2) * 6.68e-11 * ((mass_pri + mass_sec) * 1.99e30)
denom = 4 * (np.pi)**2
a = (np.cbrt(numer/denom)) / 1.50e11
```

```python
plt.scatter(a, mag_diff, s=10, color='white', edgecolor='purple', linewidth=1)
plt.xlabel('Separation [AU]', fontsize=15)
plt.ylabel(r'Magnitude Difference $[m_2 - m_1]$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
plt.savefig('/users/jess/sf_m_dwarfs/sf_plots/P_magdiff_vs_dist.png')
```
