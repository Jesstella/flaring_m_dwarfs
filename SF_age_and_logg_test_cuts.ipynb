# Testing the M (and K dwarfs) to make sure they fall as we would expect on the HR diagram. 

This script loads the final sample of G/MK binaries and plots them as a HR diagram so the user can see if any cuts need to be made. 

```python
# Import modules
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import time 
```

```python
# Pull necessary values
full_sample_array = pd.read_csv('/Users/Jess/sf_m_dwarfs/sf_data_files/G_with_binaries_with_parameters_3.csv')
sec_logg = full_sample_array['sec_logg'] # Secondary surface gravity 
sec_temp = full_sample_array['sec_temp'] # Secondary temperature
sec_age = full_sample_array['logAge'] # Secondary age (same as primary)
sec_met = full_sample_array['M_H'] # Secondary composition (same as primary)
```

```python
# Plot HR diagram WITH ages as colors
plt.figure(figsize=(10, 5))
plt.scatter(sec_temp, sec_logg, c=sec_age, cmap='cool_r', edgecolor='black')
plt.xlabel(r'$log(T_{eff})$ [K]', fontsize=15)
plt.xlim(3.73,3.3)
plt.ylim(max(sec_logg), min(sec_logg))
plt.ylabel('logg', fontsize=15)
plt.xticks(fontsize=15)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=15)
cbar.set_label('log(Age)', fontsize=15)
plt.yticks(fontsize=15)
plt.text(3.7, 4.25, 'K Dwarfs', fontsize=15)
plt.text(3.58, 4.25, 'M Dwarfs', fontsize=15)
plt.axvline(3.6, color='black')
plt.ylim(6, 4)
plt.savefig('/Users/Jess/SF_m_dwarfs/SF_plots/P_hr_companions_mk_logage.png')
```

```python
# Cut the sample by age to only include stars that are less than 1Gyr in age
age_cut = full_sample_array[(full_sample_array["logAge"] >= 9)]
sec_temp_age_cut = age_cut['sec_temp']
sec_logg_age_cut = age_cut['sec_logg']
sec_age_age_cut = age_cut['logAge']
sec_met_age_cut = age_cut['M_H']
```

```python
# Plot HR diagram WITH ages as colors, after the age cut
plt.figure(figsize=(10, 5))
plt.scatter(sec_temp_age_cut, sec_logg_age_cut, c=sec_age_age_cut, cmap='cool_r', edgecolor='black')
plt.xlabel(r'$log(T_{eff})$ [K]', fontsize=15)
plt.xlim(3.73, 3.3)
plt.ylim(max(sec_logg_age_cut), min(sec_logg_age_cut))
plt.ylabel('logg', fontsize=15)
plt.xticks(fontsize=15)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=15)
cbar.set_label('log(Age)', fontsize=15)
plt.yticks(fontsize=15)
plt.text(3.7, 4.25, 'K Dwarfs', fontsize=15)
plt.text(3.58, 4.25, 'M Dwarfs', fontsize=15)
plt.axvline(3.6, color='black')
plt.ylim(6, 4)
plt.savefig('/Users/Jess/SF_m_dwarfs/SF_plots/P_hr_companions_mk_logage_cut.png')
```

```python
# Plot HR diagram, after the cut in age; color is now metallicity
plt.figure(figsize=(10, 5))
plt.scatter(sec_temp_age_cut, sec_logg_age_cut, c=sec_met_age_cut, cmap='cool_r', edgecolor='black')
plt.xlabel(r'$log(T_{eff})$ [K]', fontsize=15)
plt.xlim(3.73, min(sec_temp_age_cut))
plt.ylim(max(sec_logg_age_cut), min(sec_logg_age_cut))
plt.ylabel('logg', fontsize=15)
plt.xticks(fontsize=15)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=15)
cbar.set_label('[M/H]', fontsize=15)
plt.yticks(fontsize=15)
plt.text(3.7, 4.25, 'K Dwarfs', fontsize=15)
plt.text(3.58, 4.25, 'M Dwarfs', fontsize=15)
plt.axvline(3.6, color='black')
plt.ylim(6, 4)
plt.savefig('/Users/Jess/SF_m_dwarfs/SF_plots/P_hr_companions_mk_logage_metal.png')
```

```python
# Cut the sample by age to only include stars that are less than 1Gyr in age and have a metallicity greater than -1.
age_comp_cut = full_sample_array[(full_sample_array["logAge"] >= 9) & (full_sample_array['M_H'] >= -1.0)]
sec_temp_age_cut = age_comp_cut['sec_temp']
sec_logg_age_cut = age_comp_cut['sec_logg']
sec_age_age_cut = age_comp_cut['logAge']
sec_comp_cut = age_comp_cut['M_H']
```

```python
# Plot HR diagram, after the cut in age.
plt.figure(figsize=(10, 5))
plt.scatter(sec_temp_age_cut, sec_logg_age_cut, c=sec_comp_cut, cmap='cool_r', edgecolor='black')
plt.xlabel(r'$log(T_{eff})$ [K]', fontsize=15)
plt.xlim(3.73, 3.3)
plt.ylim(max(sec_logg_age_cut), min(sec_logg_age_cut))
plt.ylabel('logg', fontsize=15)
plt.xticks(fontsize=15)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=15)
cbar.set_label('[M/H]', fontsize=15)
plt.yticks(fontsize=15)
plt.text(3.7, 4.25, 'K Dwarfs', fontsize=15)
plt.text(3.58, 4.25, 'M Dwarfs', fontsize=15)
plt.axvline(3.6, color='black')
plt.ylim(6, 4)
plt.savefig('/Users/Jess/SF_m_dwarfs/SF_plots/P_hr_companions_mk_logage_metal_cut.png')
```
