# Compiles all known stats for flares per star and plots them

This code will take user inputted stats and plot them for the No. of flares per star. For our multiple runs the code will use the mean flares per star and the standard deviation of the error for the plotted value.

```python
# Import modules
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# Define no. of flares per star for known quanities
m_dwarfs = pd.read_csv('/users/jess/sf_m_dwarfs/m_stars_final_stats.csv')
sf_on_stars = np.mean(m_dwarfs['sf_flares_on_stars'])
sf_error = np.mean(m_dwarfs['sf_error'])

k_dwarfs = pd.read_csv('/users/jess/sf_m_dwarfs/k_and_m_stars_final_stats.csv')
sf_on_stars_k = np.mean(k_dwarfs['sf_flares_on_stars'])
sf_error_k = np.mean(k_dwarfs['sf_error'])

s13_flares_on_stars = 5.63
s13_err = 0.366

m12_flares_on_stars = 2.47
m12_err = 0.24

s13_flares_on_sunlike = 2.316
s13_err_sunlike = 0.636

o20_flares_on_stars = 8.834
o20_err = 0.573

o20_flares_on_sunlike = 1.733
o20_err_sunlike = 0.562
```

```python
# Plot
plt.figure(figsize=(30, 8))
y = [1, 2, 3, 4, 5, 6, 7]
x = [sf_on_stars, sf_on_stars_k, s13_flares_on_sunlike, m12_flares_on_stars, s13_flares_on_stars, o20_flares_on_stars, o20_flares_on_sunlike]
errors = [sf_error, sf_error_k, s13_err_sunlike, m12_err, s13_err, o20_err, o20_err_sunlike]

plt.scatter(x, y, s=250, color=['#a6cee3', '#b2df8a', '#fb9a99', '#fdbf6f', '#cab2d6', '#bdbdbd', '#c994c7'], edgecolor=['#1f78b4', '#33a02c', '#e31a1c', '#ff7f00', '#6a3d9a', '#000000', '#ce1256'])
plt.errorbar(x, y, xerr=errors, ls='None', color=['#1f78b4', '#33a02c', '#e31a1c', '#ff7f00', '#6a3d9a', '#000000', '#ce1256'], linewidth=4)
plt.xticks(fontsize=25)
plt.xlabel('No. of Flares per Star Showing Flares', fontsize=25)
plt.yticks(ticks=[1, 2, 3, 4, 5, 6, 7], labels=['This work\nM dwarfs', 'This work\nM and K dwarfs', 'Shibayama (2013)\n Sun-like', 'Maehara (2012)', 'Shibayama (2013)', 'Okamoto (2020)', 'Okamoto (2020)\n Sun-like'], fontsize=25)
plt.savefig('/Users/jess/sf_m_dwarfs/sf_plots/P_final_stats_all.png')
plt.show()
```
